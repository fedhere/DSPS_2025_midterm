Side note: I know I promised examples in other areas of physics than astro, but astrophysicists have really adopted open science and reproducibility practices beyond other areas of physics and there are few open curated datasets to support analysis suitable for this level work.

10 stars explode every second in the sky, most of them unseen as they are too far for their light to reach us and definitely too far to see the star before the explosion. Stars collapse and explode, merge and explode, and stars of different nature lead to different explosion properties. We often play a game stellar foresics, trying to figure out what star and why blew off from the explosion properties.

We see emission from stellar explosions in different energy regimes from different instruments and observatories.
A new observatory, the Einstein Probe, opened an observational window in the soft X-ray dynamic sky, enabling discoveries of rapid X-ray transients.

Long duration GRBs, which are understood to be explosions and mergers of massive stars, are expected to emit in X-ray, but only a fraction of the transients detected by EP were associated with Gamma Ray emission (as observed by other telescopes from the same region of the sky). Sometimes we do not detect the GRB because we are not looking in the right place, or they are too faint. This raises the question: are these soft, rapid X-ray emissions simply a manifestation of long duration Gamma Ray Bursts in X-ray wavelength? or are they different transients altogether? _That would be exciting! we may have discovered a new physical phenomenon! but how do we make sure?_

Connor et al. 2025 [https://arxiv.org/pdf/2509.07141#page=8.87] approached the problem this way: they found a comprehensive datasets of long duration Gamma Ray Bursts (Connor et al. 2022 https://academic.oup.com/mnras/article/515/4/4890/6649214) and a dataset of short Gamma Ray Bursts (https://www.mpe.mpg.de/~jcg/grbgen.html), a different less well understood phenomenon. They compare the spatial distribution, namely the distribution of of redshifts, which means distances from us (because all these phenomena are, as far as we can tell, randomely distributed on the plane of the sky), for these three samples and ask the question: can they be the same? and most importantly, are the distributions of EP transients with GRB and without GRB both the same as the distribution of long GRBs? if they are, it is possible we are just missing the GRB detections, but the gamma ray emission could be there for all transients detected by the EP. But if they are not, then these are two different phenomena and there may be new physics!!!
