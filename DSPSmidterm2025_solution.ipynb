**Extra Credit** if you have chosen the right statistical test, there is a plot that naturally goes with it and helps you visualize your findings. Add this plot (as seen in some homework)



```python
# This cell is intentionally left empty as a placeholder for potential future code related to the KS test visualization.
# The visualization code is included within the getKSpvalue function in the next cell.
```

```python
def getKSpvalue(samp1, samp2, s1name, s2name):
  ksresult = ks_2samp(samp1["Redshift"], samp2["Redshift"])[1]
  cancannot = "can" if ksresult < 0.05 else "cannot"
  print(f"We {cancannot} reject the null hypothesis that the samples are the same")
  ax = samp1.hist(cumulative=True, density=True, histtype='step', label=s1name)
  samp2.hist(cumulative=True, density=True, histtype='step', ax=ax, label=s2name)
  plt.legend()
  return ksresult


for i,s1 in enumerate(samples[:-2]):
  for j,s2 in enumerate(samples[i+1:-2]):
    print(seriesnames[i], seriesnames[i+j+1])
    print(f"p-value {getKSpvalue(s1,s2,seriesnames[i], seriesnames[i+j+1]):.2f}")
```

We can rule out the hypothesis that shortGRB redshift distribution is consistent with either longGRB or the EP soft Xray transients. Short GRB are a distinct population (p-value > 0.05)

However, we cannot rule out that the soft X-ray EP transients redshift distribution is identical to the longGRB's one, implying that the EP transients may be the same class of objects as longGRBs.
